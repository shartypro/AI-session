```
import cv2
import mediapipe as mp

mp_hands = mp.solution.hands
mp_drawing =  mp.solutions.drawing_utils
hands = mp_hands.Hands()

cap = mcv2.videocapture(0)

while True :
  ret ,frame = cap.read()

  if not ret :
    print("Camera isNOT workintg")
    break

  frame = cv2.flip(frame ,1)

  h,w,_ = frame.shape

  #RGB , BGR
  # R = 255
  #G = 255
  #B = 255
  # (255 , 0 , 0)
  frame_rg = cv2.cvtcolor(frame , cv2.COLOR_BGR2RGB)

  results = hands.process(frame_rgb)

  if results.multi_han_landmarks :
    for hand_landmarks in results.multi_hand_landmarks :

      mp_drawing.draw_landmarks(frame , hand land_landmarks , mp_hands.HAND_CONNECTIONS)

      cx = int(hand_landmarks.landmark[9].x * w)
      cy = int(hand_landmarks.landmark[9].y * h)


      cv2.circle(frame , (cx , cy) , 20 , (255 , 0 ,0), -1)

cap.release()
cv2.destroyALLwindows()
```
